```python
version = 3.0

class JuliaSet():
    def __init__(self, c, n=100):
        self.c = c
        self.n = n
        self._d = .001
        self.complexplane = []
        self.set = []
    
    def juliamap(self, z):
        return z**2 + self.c
    
    def iterate(self, z):
        m = 0
        while True:
            z = self.juliamap(z)
            m +=1
            if abs(z)>2:
                return (m)
            if m>=self.n:
                return 0
                    
    def my_range(self, start, end, step):
        while start <= end:
            yield start
            start += step
                
    def gen_complexplane(self):
        self._complexplane = []
        for x in self.my_range(-2, 2, self._d):
            for y in self.my_range(-2, 2, self._d):
                self._complexplane.append(complex(x,y))

    def set_spacing(self, d):
        self._d = d
        self.gen_complexplane()
        #print('spacing is now ' + repr(self._d) )
        
    def generate(self):
        self.set = [self.iterate(z) for z in self._complexplane]
        return self.set
    
js = JuliaSet(complex(0,0),10)
js.iterate(complex(0,0))
js.set_spacing(.1)
js.generate()
#print("Testing the numbers...")
#print(js.c)
#print(js.n)
#print(js._d)
#print("Testing juliamap...")
#print(js.juliamap(complex(3,2)))
#print("Testing Iterate...")
#print("Testing set_spacing...")
#print("Testing ")
```
